```python
import pandas as pd
import numpy as np
import openpyxl
from tabulate import tabulate
```

```python
df = pd.read_excel('data/materials.xlsx', index_col=0)
df = df.sort_values(by='mass').reset_index()
df
```

```python
# assign data
mydata = zip(df['fomula'].values, df['mass'].values, df['abundance'].values)

# create header
head = ['fomular', 'mass', 'abundance']

# display table
print(tabulate(mydata, headers=head, tablefmt="grid"))
```
